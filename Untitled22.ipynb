```python
import pandas as pd
auto_mpg = pd.read_csv("/auto-mpg.csv")
auto_mpg.head()
```

```python
auto_mpg = auto_mpg[auto_mpg.horsepower != '?']
#auto_mpg.loc[:,'horsepower'] = auto_mpg.horsepower.astype('float') # changes only the values to float
auto_mpg = auto_mpg.astype({'horsepower': float}) # changes the values to float and the schema of the dataframe (recommended)
auto_mpg.dtypes
```

```python
#Visualise Data
auto_mpg.describe()
```

```python
_ = auto_mpg.hist(figsize=(20, 14))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

selected_columns = ['mpg', 'displacement', 'weight', 'acceleration']
sns.pairplot(auto_mpg, vars=selected_columns)
plt.show()
```

```python
sns.boxplot(x='model year', y='mpg', data=auto_mpg)
```
